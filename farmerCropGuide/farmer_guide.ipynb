```python
# KMeans  accuracy normalize
from sklearn.preprocessing import MinMaxScaler, RobustScaler,StandardScaler,Normalizer
std = StandardScaler()


x_transformed  = std.fit_transform(x)
```

```python
x_transformed
```

```python
x_transformed.shape
```

```python
from sklearn.cluster import KMeans 
kmeans = KMeans()
kmeans.fit(x_transformed)
```

```python
np.unique(kmeans.labels_)
```

```python
kmeans.labels_.shape , df.shape
```

```python
df['cluster_no'] = kmeans.labels_
```

```python
groupno_five = df[df['cluster_no']==5]
```

```python
groupno_five['label'].unique()
```

```python
import os, joblib 
os.makedirs('models',exist_ok=True)

joblib.dump(kmeans,'./models/kmeans_model.lb')
joblib.dump(std,'./models/standardScalar.lb')
print('Your models has been saved!')
```

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_score(x_transformed,kmeans.labels_)
```
