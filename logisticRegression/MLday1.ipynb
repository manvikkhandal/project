```python
# Get the brands that appear at least 10 times
bike_brand = list(df['brand'].value_counts().head(16).index)
bike_brand
```

```python
df2 = df[df['brand'].isin(bike_brand)]
df2.head()
```

```python
df2['brand'] = df2['brand'].map(br) # label encoding
df2.head()
```

<h1>ML-Day2</h1>

```python
dt2 = {}
for i in range(len(df2['brand'].value_counts())):
    key = df2['brand'].value_counts().keys()[i]
    value = i + 1
    dt2[key] = value
```

splitting data independent and dependent variables


```python
df2.drop(['bike_name','city'],axis=1,inplace=True)
```

```python
df2.head()
```

```python
# Supervised learning   y--> price remaining x
x = df2.drop('price',axis=1) # independent variable
x
```

```python
y = df2[['price']] # dependent variable
y
```

Importing train test split


```python
from sklearn.model_selection import train_test_split
```

```python
x_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.2)  # test size is 20%
```

```python
x_train,X_test,y_train,y_test
```

Importing Linear Regression algorithm class

```python
from sklearn.linear_model import LinearRegression
```

```python
# instance creation
lnr = LinearRegression()
```

```python
## algorithm train
lnr.fit(x_train,y_train)
```

```python
## model evaluation 
print("Training score: ",lnr.score(x_train,y_train)*100)
print("Testing score : ", lnr.score(X_test,y_test)*100)
```

```python
### prediction or testing phase
prediction = lnr.predict(X_test)
```

```python
prediction.shape
```

```python
comparedf = pd.concat([X_test,y_test],axis=1) 
## combined the X_test and y_test
```

```python
comparedf['prediction'] = prediction
```

```python
comparedf.head(40)
```

```python
comparedf['residual'] = comparedf['price'] - comparedf['prediction']
```
