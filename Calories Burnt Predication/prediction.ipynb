<h1>Calories Burnt Prediction</h1>

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
import pickle
```

```python
df = pd.read_csv("calories.csv")
df
```

```python
df1 = pd.read_csv("exercise.csv")
df1
```

```python
df2 = pd.concat((df,df1),axis=1)
df2
```

```python
df2.drop("User_ID",inplace=True,axis=1)
```

```python
df2.head()
```

```python
df2.info()
```

```python
df2.Gender.value_counts()
```

```python
gender_maping = {"female":0,"male":1}
```

```python
df2['Gender'] = df2['Gender'].map(gender_maping)
```

```python
df2.head()
```

```python
x = df2.drop("Calories",axis=1)
y = df2.Calories
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
lr_model = LinearRegression()
dt_model = DecisionTreeRegressor()
rf_model = RandomForestRegressor()
```

```python
lr_model.fit(X_train, y_train)
dt_model.fit(X_train, y_train)
rf_model.fit(X_train, y_train)
```

```python
print("LinearRegression training score:", lr_model.score(X_train,y_train))
print("LinearRegression testing score:", lr_model.score(X_test,y_test))
```

```python
print("descisiontree training score:", dt_model.score(X_train,y_train))
print("descisiontree testing score:", dt_model.score(X_test,y_test))
```

```python
print("randomforest training score:", rf_model.score(X_train,y_train))
print("randomforest testing score:", rf_model.score(X_test,y_test))
```
