```python
#write mode can also create a file if not exist in directory
message = """
HII we are form upflairs, and thank you for joining our data
science and machine program.
"""
file = open('demo.txt','w')
file.write(message)
file.close()
```

```python
file = open("demo.txt",'r')
data = file.read()
print(data)
```

```python
message = """
\nHII we are form upflairs, and thank you for joining our data
science and machine program.
"""
file = open('demo.txt','a')
file.write(message)
file.close()
```

```python
file = open("demo.txt",'r')
data = file.read()
file.close()
print(data)
```

```python
with open("demo.txt",'r') as file:
    data = file.read()
    print(data)
```
